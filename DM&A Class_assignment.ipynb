# JP Morgan Chase & Co

```python
JPM_weekly = get_data("JPM", start_date="08/12/2017", end_date="08/12/2019", index_as_date=True, interval="1wk")
JPM_weekly.reset_index(inplace=True)  #Reseting the index of DataFrame
JPM_weekly.rename(columns={'index': 'timestamp'}, inplace=True)   #Renaming the index column of the DataFrame to timestamp
JPM_weekly
```

```python
ohlc2_data = JPM_weekly[['timestamp', 'open', 'high', 'low', 'close']]  # Select columns 'timestamp', 'open', 'high', 'low', and 'close' from the DataFrame JPM_weekly
ohlc2_data.set_index('timestamp', inplace=True)   # Set the 'timestamp' column as the index of the DataFrame ohlc2_data

# Creating a candlestick chart
mpf.plot(ohlc2_data, type='candle', title='Weekly Chart', ylabel='Price', style='classic')
```

```python
import matplotlib.pyplot as plt

# Assuming you have the JPM_weekly DataFrame

# Create a figure and axis for the plot
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the 'open' prices in blue
ax.plot(JPM_weekly['timestamp'], JPM_weekly['open'], label='Open', color='blue')

# Plot the 'close' prices in red
ax.plot(JPM_weekly['timestamp'], JPM_weekly['close'], label='Close', color='red')

# Set labels and title
ax.set_xlabel('Dates')
ax.set_ylabel('Price')
ax.set_title('JPM Weekly Open and Close Prices')

# Add a legend
ax.legend()

# Show the plot
plt.xticks(rotation=45)
plt.show()
```

```python
# Calculate the correlation matrix
correlation_matrix_1 = JPM_weekly.corr()

# Create a heatmap of the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix_1, annot=True, cmap='cool', linewidths=.5, fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```
